```python
train_y = df['Survived']
```

```python
test_df = pd.read_csv('test.csv')
```

```python
test_X = test_df[['Sex', 'Pclass']]
```

```python
test_X = pd.get_dummies(test_X, columns=['Sex'])
```

```python
test_X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(train_X, train_y)
```

```python
prediction = model.predict(test_X)
```

```python
prediction
```

```python
submission_df = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': prediction})
```
